```python
import pandas as pd
import numpy as np
station = pd.DataFrame.from_csv("stations-i880n.csv")
station = station[["ID", "Abs PM", "Length"]]
display(station.head())
display(station.describe())
```

```python
incident = pd.read_csv("otherIncident.csv", error_bad_lines=False)
incident = incident[["Start_Time", "Duration_(mins)", "Abs_PM"]]
incident["ID1"] = '0'
incident["ID2"] = '0'
incident["ID3"] = '0'
display(incident.head())
display(incident.describe())
```

```python
counter = 0
for index1, row1 in incident.iterrows():
    for index2, row2 in station.iterrows():
        if(row1["Abs_PM"] >= row2["Abs PM"] and row1["Abs_PM"] <= row2["Abs PM"]+row2["Length"] and counter ==0):
            incident.at[index1,'ID1'] = row2["ID"]
            #counter = counter + 1
            continue
        '''if(row1["Abs_PM"] >= row2["Abs PM"] and row1["Abs_PM"] <= row2["Abs PM"]+row2["Length"] and counter ==1):
            incident.at[index1,'ID2'] = row2["ID"]
            counter = counter +1
            continue
        if(row1["Abs_PM"] >= row2["Abs PM"] and row1["Abs_PM"] <= row2["Abs PM"]+row2["Length"] and counter ==2):
            incident.at[index1,'ID3'] = row2["ID"]
            counter = counter +1
            continue'''
    #counter = 0
display(incident.head())
display(incident.describe())
```

```python
from datetime import datetime as dt
dtobject = dt.strptime(incident.iloc[0]["Start_Time"],"%m/%d/%y %H:%M")
print(dtobject.minute)
```

```python
for index, row in incident.iterrows():
    incident.at[index, 'Start_Time'] = dt.strptime(incident.at[index, "Start_Time"],"%m/%d/%y %H:%M")
display(incident.head())
```

```python
occlist = []
for i in range(1,4):
    df = pd.read_csv("occ"+str(i)+".csv", error_bad_lines=False)
    display(df.head())
    df["ID"] = df["Station ID"]
    df = df[["5 Minutes", "Occupancy (%)", "ID"]]
    occlist.append(df)
occupancy = pd.concat(occlist)
#occupancy = occupancy[["5 Minutes", "Occupancy (%)"]]
display(occupancy.head())
```

```python
display(occupancy.describe())
```

```python
flowlist = []
for i in range(1,4):
    df = pd.read_csv("flow_speed"+str(i)+".csv", error_bad_lines=False)
    df["ID"] = df["Station ID"]
    df = df[["5 Minutes", "Flow (Veh/5 Minutes)", "Speed (mph)","ID"]]
    
    flowlist.append(df)
flow = pd.concat(flowlist)
#flow = pd.DataFrame.from_csv("flow_speed.csv")
#flow = flow[["5 Minutes", "Flow (Veh/5 Minutes)", "Speed (mph)"]]
display(flow.head())
display(flow.describe())
```

```python
flowSpeedOcc = pd.merge(occupancy, flow)
flowSpeedOcc = flowSpeedOcc[["ID","5 Minutes", "Flow (Veh/5 Minutes)", "Speed (mph)", "Occupancy (%)"]]
display(flowSpeedOcc.head())
display(flowSpeedOcc.describe())
```

```python
for index, row in flowSpeedOcc.iterrows():
    flowSpeedOcc.at[index, '5 Minutes'] = dt.strptime(flowSpeedOcc.at[index, "5 Minutes"],"%m/%d/%Y %H:%M")
```

```python
flowSpeedOcc["Incident"] = 0
display(flowSpeedOcc.head())
```
